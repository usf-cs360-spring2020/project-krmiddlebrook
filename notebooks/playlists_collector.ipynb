```python
playlists_arr = []

# collect the sounds of Spotify playlists
playlists = spm.client().user_playlists('thesoundsofspotify')
pprint(playlists)
bar = tqdm(total=playlists.get('total', 0))
while playlists:
    playlists_arr.extend(playlists.get('items', []))
    if playlists['next']:
        playlists = spm.client().next(playlists)
    else:
        playlists = None
    bar.update(50)
```

```python
# remove non-genre specific playlists
print(playlists_df.name.str.contains('The Sound of').value_counts())
playlists_df = playlists_df.loc[playlists_df.name.str.contains('The Sound of')==True, :].copy().reset_index(drop=True)
```

```python
from country_list import countries_for_language
# countries_for_language returns a list of tuples now, might be changed to an OrderedDict
countries = dict(countries_for_language('en'))
print(len(countries))
pprint(countries)

for k,v in countries.items():
    countries[k] = v.replace('.', '')
pprint(countries)
    
```

```python
# remove as many country specific genres as is easily possible

# remove country codes
print(playlists_df.name.apply(lambda x: any(key in x for key in countries.keys())).value_counts())
playlists_df = playlists_df.loc[playlists_df.name.apply(lambda x: any(key in x for key in countries))==False, :].copy().reset_index(drop=True)
print(playlists_df.shape)

# remove country names
print(playlists_df.name.apply(lambda x: any(key in x for key in countries.values())).value_counts())
playlists_df = playlists_df.loc[playlists_df.name.apply(lambda x: any(key in x for key in countries))==False, :].copy().reset_index(drop=True)
playlists_df.shape
```

```python
# remove the one song from every genre playlist
playlists_df = playlists_df.loc[1:, :].reset_index(drop=True)
playlists_df.sample(n=10)
```

```python
# collect the track data for each playlist
playlist_tracks_arr = []
fields="next,total,items(added_at,added_by.id, track(artists(id, name), album.id,album.name, album.album_type, album.release_date, id, name, popularity, preview_url, external_ids.isrc, explicit))"

playlist_ids = playlists_df.id.dropna().unique()


# outer loop of all playlists
for playlist_id in tqdm(playlist_ids):
    playlist_tracks = spm.client().playlist_tracks(playlist_id, fields=fields, market='US')
#     pprint(playlist_tracks)
    
    # individual playlist loop to get all associated tracks
    with tqdm(total=playlist_tracks.get('total', 0)) as pbar:
        while playlist_tracks:
            for track in playlist_tracks.get('items', []):
                track['playlist_id'] = playlist_id

            playlist_tracks_arr.extend(playlist_tracks.get('items', []))
            if playlist_tracks['next']:
                playlist_tracks = spm.client().next(playlist_tracks)
            else:
                playlist_tracks = None
            pbar.update(100)
```

```python
def unpack_playlist_item(item):
    """Unpacks a playlist track object (e.g., dict) into a cleaner representation 
    
    """
    
    unpacked = {
        'playlist_id': item.get('playlist_id', None),
        'added_at': item.get('added_at', None),
        'added_by_id': item.get('added_by', {'id': None}).get('id'),
    }
    
    if item.get('added_by', None):
        del item['added_by']
    
    try:
        # track section
        track = item.get('track', None)
        if track:
            # track.album section 
            album = track.get('album', None)
            if album:
                unpacked['album_type'] = album.get('album_type', None)
                unpacked['album_id'] = album.get('id', None)
                unpacked['album_name'] = album.get('name', None)
                unpacked['release_date'] = album.get('release_date', None)
                del track['album'] # remove album section from track dict

            # unpack track isrc 
            unpacked['isrc'] = None
            unpacked['isrc'] = track.get('external_ids', {'isrc': None}).get('isrc')
            if track.get('external_ids', None):
                del track['external_ids']

            # add track (with unpacked album metadata and isrc and list of packed artists metadata) to unnpacked dict
            unpacked.update(track)
    except:
        pass
    
    return unpacked
```
